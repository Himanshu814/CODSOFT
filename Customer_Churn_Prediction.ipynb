#  **Customer Churn Prediction**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, roc_curve, auc

import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv('/content/Churn_Modelling.csv')
df
```

```python
#delete unneeded columns
df.drop(columns=['RowNumber','CustomerId','Surname'],inplace=True)
df
```

```python
# number of null values
df.isnull().sum()
```

```python
# number of duplicate rows/entries
df.duplicated().sum()
```

**Feature Encoding**

```python
X=df.drop('Exited', axis=1)
y=df['Exited']
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder,LabelEncoder,StandardScaler
ct=ColumnTransformer(
    [
        ('ohe',OneHotEncoder(drop='first'),['Geography','Gender']),
        ('sc',StandardScaler(),['CreditScore','Age','Balance','EstimatedSalary']),
    ]
)
X_new = ct.fit_transform(X)
X_new
```

**Training and testing data**

```python
# Train test split
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
X_train, X_test, y_train, y_test = train_test_split(X_new, y, test_size=0.2)
```

```python
# testing out different models
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
```

**Logistic Regression**

```python
# Logistic Regression
model_lr = LogisticRegression()
model_lr.fit(X_train, y_train)
y_pred = model_lr.predict(X_test)
LR_accuracy = accuracy_score(y_test, y_pred)
LR_accuracy
```

**Random Forest**

```python
# Random Forest
model_rf = RandomForestClassifier()
model_rf.fit(X_train, y_train)
y_pred = model_rf.predict(X_test)
RF_accuracy = accuracy_score(y_test, y_pred)
RF_accuracy
```

**Gradient Boosting Classifier**

```python
# Gradient Classifier
model_gc = GradientBoostingClassifier()
model_gc.fit(X_train, y_train)
y_pred = model_gc.predict(X_test)
GB_accuracy = accuracy_score(y_test, y_pred)
GB_accuracy
```

Comparison of Gradient Boosting,
RandomForest and Logistic Regression

```python
Algorithms = ['Gradient Boosting','RandomForest', 'Logistic Regression']
accuracy = [GB_accuracy, RF_accuracy, LR_accuracy]

FinalResult=pd.DataFrame({'Algorithm':Algorithms, 'Accuracy':accuracy})

FinalResult
```

```python
plt.figure(figsize=(7, 5))
plt.bar(FinalResult['Algorithm'], FinalResult['Accuracy'], color='pink')
plt.xlabel('Algorithm')
plt.ylabel('Accuracy')
plt.title('Accuracy of Different Algorithms')
plt.ylim(0, 1)  # Set the limit of y-axis from 0 to 1 (accuracy ranges from 0 to 1)
plt.xticks(rotation=45)  # Rotate x-axis labels for better visibility
plt.grid(axis='x')  # Add gridlines only along the x-axis
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```
